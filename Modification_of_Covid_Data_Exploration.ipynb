# 1) Importing Library

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

#2). Load your dataset


````python
health_df = pd.read_csv('dcov.csv')
````

#3). Display the some first rows of the dataset


````python
print("First few rows of the health dataset:")
print(health_df.head())
````

#4). Summary statistics


````python
print("\nSummary statistics:")
print(health_df.describe())
````

#5). Check for missing values


````python
print("\nMissing values:")
print(health_df.isnull().sum())
````

#6). Visualize the distribution of the target variable 'class'


````python
plt.figure(figsize=(6, 4))
sns.countplot(x='class', data=health_df)
plt.title('Distribution of Classes')
plt.show()
````

#7).correlation matrix


````python
correlation_matrix = health_df.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
````

#8).distribution by class Visualisation of Data Using Graphs


````python
for feature in health_df.columns[:-1]:
    plt.figure(figsize=(20, 6))
    sns.histplot(x=feature, hue='class', data=health_df, kde=True)
    plt.title(f'Distribution of {feature} by Class')
    plt.show()
````

#9).Box plots for selected features


````python
selected_features = ['Duration', 'Incoming_calls', 'outgoing_calls', 'incoming_messages', 'Active_time', 'posts', 'nervous', 'sleephr', 'walkinghr', 'exercisehr', 'idlehr', 'heart_rate']
for feature in selected_features:
    plt.figure(figsize=(20, 6))
    sns.boxplot(x='class', y=feature, data=health_df)
    plt.title(f'Box Plot of {feature} by Class')
    plt.show()
````

#10). Distribution Of classes using countplot


````python
categorical_features = ['Age', 'Duration', 'heart_rate', 'idlehr', 'sleep_pattern']
for feature in categorical_features:
    plt.figure(figsize=(20, 6))
    sns.countplot(x=feature, hue='class', data=health_df)
    plt.title(f'Distribution of {feature} by Class')
    plt.show()
````

#11). Correlation Analysis with Target Variable:


````python
correlations_with_target = health_df.corr()['class'].sort_values(ascending=False)
print("Correlations with the target variable:")
print(correlations_with_target)
````
